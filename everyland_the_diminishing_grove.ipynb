````python
import glob

import geopandas as gpd
import numpy as np
import pandas as pd
from sklearn import preprocessing
````

````python
FILTERED_YEAR = np.arange(2000, 2021, 1)
````

### Land & Forest Data

Datasets sources list: No.1 & No.2

````python
land = pd.read_csv(
    "datasets/land_undata/SYB64_145_202110_Land.csv",
    encoding='windows_1258',
    skiprows=1,
)
land.rename(columns={'Unnamed: 1': 'Country', 'Value': 'Land Area'}, inplace=True)
land.drop(columns={'Region/Country/Area', 'Footnotes', 'Source'}, inplace=True)
land = land.pivot(
    index=['Country', 'Year'], columns=['Series'], values=['Land Area']
).reset_index()
land_columns_name = [
    'Country',
    'Year',
    '% Arable Area',
    'Total Arable Area (hectares)',
    '% Forest Cover Area',
    'Total Forest Cover Area (hectares)',
    '% Important sites for terrestrial biodiversity protected',
    'Total Land Area (hectares)',
    '% Permanent Crops Area',
    'Total Permanent Crops Area (hectares)',
]
land.columns = [''] * len(land.columns)
land.columns = land_columns_name
land = land[land['Year'].isin(FILTERED_YEAR)]
land.replace(',', '', regex=True, inplace=True)
land = land.set_index(['Country']).apply(pd.to_numeric).reset_index()


forest = pd.read_csv(
    "https://raw.githubusercontent.com/owid/owid-datasets/master/datasets/Forest%20land%2C%20deforestation%20and%20change%20(FAO%2C%202020)/Forest%20land%2C%20deforestation%20and%20change%20(FAO%2C%202020).csv"
)
forest = forest[forest['Year'].isin(FILTERED_YEAR)]
# forest = forest.dropna().reset_index(drop=True)
forest = forest[
    [
        'Entity',
        'Year',
        'Deforestation',
        'Deforestation as share of forest area',
        'Forest area',
        'Share of global deforestation',
        'Share of global forest area',
        'Forest cover',
    ]
]
forest.rename(
    columns={
        'Entity': 'Country',
        'Deforestation': 'Total Deforestation Area (hectares)',
        'Deforestation as share of forest area': '% Deforestation',
        'Forest area': 'Total Forest Cover Area (hectares)',
        'Share of global deforestation': '% Share of Global Deforestation',
        'Share of global forest area': '% Share of Global Forest Cover Area',
        'Forest cover': '% Forest Cover Area',
    },
    inplace=True,
)
forest = forest.set_index(['Country']).apply(pd.to_numeric).reset_index()


# display(land[~land['Country'].isin(forest['Country'].unique())]['Country'].unique()) # land unique countries
# display(forest[~forest['Country'].isin(land['Country'].unique())]['Country'].unique()) # forest unique countries

country_alignment_map_1 = {
    'Bolivia (Plurin. State of)': 'Bolivia',
    'Brunei Darussalam': 'Brunei',
    'Dem. People's Rep. Korea': 'North Korea',
    'Dem. Rep. of the Congo': 'Democratic Republic of Congo',
    'Falkland Islands (Malvinas)': 'Falkland Islands',
    'Iran (Islamic Republic of)': 'Iran',
    'Lao People's Dem. Rep.': 'Laos',
    'Micronesia': 'Micronesia (Fed. States of)',
    'North Macedonia': 'Macedonia',
    'Republic of Korea': 'South Korea',
    'Republic of Moldova': 'Moldova',
    'Russian Federation': 'Russia',
    'State of Palestine': 'Palestine',
    'Syrian Arab Republic': 'Syria',
    'United Rep. of Tanzania': 'Tanzania',
    'United States of America': 'United States',
    'Venezuela (Boliv. Rep. of)': 'Venezuela',
    'Viet Nam': 'Vietnam',
}

country_alignment_map_2 = {
    'Bonaire, Sint Eustatius and Saba': 'Bonaire St. Eustatius & Saba',
    'Cape Verde': 'Cabo Verde',
    'Czech Republic': 'Czechia',
    'French Guyana': 'French Guiana',
    'Saint Vincent and the Grenadines': 'Saint Vincent & Grenadines',
    'Sint Maarten (French part)': 'Saint Martin (French part)',
    'Timor': 'Timor-Leste',
}

land = land[land['Country'] != 'Micronesia']
land['Country'].replace(country_alignment_map_1, inplace=True)
land.to_csv("datasets/processed/land.csv", index=False)
forest['Country'].replace(country_alignment_map_2, inplace=True)
forest.to_csv("datasets/processed/forest.csv", index=False)

dataset = pd.merge(
    forest,
    land.drop(columns=['% Forest Cover Area', 'Total Forest Cover Area (hectares)']),
    how='outer',
    on=['Country', 'Year'],
)

dataset = dataset.reindex(
    columns=[
        'Country',
        'Year',
        'Total Deforestation Area (hectares)',
        '% Deforestation',
        '% Share of Global Deforestation',
        'Total Forest Cover Area (hectares)',
        '% Forest Cover Area',
        '% Share of Global Forest Cover Area',
        'Total Arable Area (hectares)',
        '% Arable Area',
        'Total Permanent Crops Area (hectares)',
        '% Permanent Crops Area',
        'Total Land Area (hectares)',
        '% Important sites for terrestrial biodiversity protected',
    ]
)

dataset_alignment = {
    'China Hong Kong SAR': 'Hong Kong',
    'China Macao SAR': 'Macao',
    'Netherlands Antilles [former]': 'Netherlands Antilles',
}
dataset['Country'].replace(dataset_alignment, inplace=True)
````

### Extra Land

Datasets sources list: No.6

````python
ail = pd.read_csv(
    "datasets/agricultural_irrigated_land_worldbank/API_AG.LND.IRIG.AG.ZS_DS2_en_csv_v2_3480623.csv",
    skiprows=4,
)
ail = pd.melt(ail, id_vars=['Country Name', 'Country Code'], value_vars=ail.columns[4:])
ail.rename(
    columns={'variable': 'Year', 'value': '% Agricultural Irrigated Land'}, inplace=True
)

al = pd.read_csv(
    "datasets/agricultural_land_worldbank/API_AG.LND.AGRI.ZS_DS2_en_csv_v2_3470479.csv",
    skiprows=4,
)
al = pd.melt(al, id_vars=['Country Name', 'Country Code'], value_vars=al.columns[4:])
al.rename(columns={'variable': 'Year', 'value': '% Agricultural Land'}, inplace=True)

arl = pd.read_csv(
    "datasets/arable_land_worldbank/API_AG.LND.ARBL.ZS_DS2_en_csv_v2_3475572.csv",
    skiprows=4,
)
arl = pd.melt(arl, id_vars=['Country Name', 'Country Code'], value_vars=arl.columns[4:])
arl.rename(columns={'variable': 'Year', 'value': '% Arable Land'}, inplace=True)

exland = pd.merge(ail, al, on=['Country Name', 'Country Code', 'Year'])
exland = pd.merge(exland, arl, on=['Country Name', 'Country Code', 'Year'])
exland = exland[~exland['Year'].str.contains('Unnamed: 65')]
exland['Year'] = exland['Year'].astype(int)
exland = exland[exland['Year'].isin(FILTERED_YEAR)]
exland = (
    exland.rename(columns={'Country Name': 'Country'})
    .reset_index(drop=True)
    .drop(columns='Country Code')
)

exland.to_csv("datasets/processed/exland.csv", index=False)
````

### Agriculture Data

Datasets sources list: No.3

````python
# Agricultural productivity is total factor productivity (TFP) where TFP measures the amount of agricultural output
# produced from the combined set of land, labor, capital, and material resources employed in farm production.
agricultural_productivity = pd.read_csv(
    "https://raw.githubusercontent.com/owid/owid-datasets/master/datasets/Agricultural%20total%20factor%20productivity%20(USDA)/Agricultural%20total%20factor%20productivity%20(USDA).csv"
)
agricultural_productivity.drop(
    columns=[
        'ag_land_index',
        'labor_index',
        'capital_index',
        'materials_index',
        'output_quantity',
    ],
    inplace=True,
)
agricultural_productivity.columns = [
    'Country',
    'Year',
    'TFP',
    'Productivity Outputs',
    'Productivity Inputs',
    'Crop Output Quantity',
    'Animal Output Quantity',
    'Fish Output Quantity',
    'Agricultural Land Quantity',
    'Labor Quantity',
    'Capital Quantity',
    'Machinery Quantity',
    'Livestock Quantity',
    'Fertilizer Quantity',
    'Animal Feed Quantity',
    'Cropland Quantity',
    'Pasture Quantity',
    'Irrigation Quantity',
]
agricultural_productivity = agricultural_productivity[
    agricultural_productivity['Year'].isin(FILTERED_YEAR)
]

# display(agricultural_productivity[~agricultural_productivity['Country'].isin(dataset['Country'].unique())]['Country'].unique())
agricultural_productivity['Country'].replace({'Timor': 'Timor-Leste'}, inplace=True)
agricultural_productivity.to_csv(
    "datasets/processed/agricultural_productivity.csv", index=False
)

dataset = pd.merge(
    dataset, agricultural_productivity, how='outer', on=['Country', 'Year']
)
````

### Trade Export & Import Data

Datasets sources list: No.4

````python
trade = list()
target_file = glob.glob("datasets/trade_wits/Trade_*.csv")
target_file = [str(f).replace('\\', '/') for f in target_file]
# country_metadata = pd.read_excel("http://wits.worldbank.org/data/public/AllCountries.xls")
# country_metadata.to_csv("datasets/trade_wits/country_metadata.csv", index=False)
country_metadata = pd.read_csv("datasets/trade_wits/country_metadata.csv")
iso_to_country = dict(zip(country_metadata['ISO3'], country_metadata['Country Name']))
# Retrieve from WITS lOGIN -> Support Materials -> Product Nomenclature -> GTAP, Sections -> Download
product_metadata = pd.read_csv("datasets/trade_wits/product_metadata.csv")
code_to_desc = dict(
    zip(
        product_metadata['GTAP Product Code'],
        product_metadata['GTAP Product Description'],
    )
)
# quantity_token_metadata = pd.read_excel("http://wits.worldbank.org/data/public/UNCOMTRADEQuantityCodes.xls")
# quantity_token_metadata.to_csv("datasets/trade_wits/quantity_token_metadata.csv", index=False)
quantity_token_metadata = pd.read_csv("datasets/trade_wits/quantity_token_metadata.csv")
token_to_unit = dict(
    zip(quantity_token_metadata['Qty Token'], quantity_token_metadata['Short Name'])
)

for file in target_file:
    c = pd.read_csv(file)
    c = c.groupby(['ReporterISO3', 'ProductCode', 'QuantityToken'], as_index=False)[
        ['TradeValue', 'Quantity', 'NetWeight']
    ].sum()
    c['ReporterISO3'] = c['ReporterISO3'].map(iso_to_country)
    c['ProductCode'] = c['ProductCode'].map(code_to_desc)
    c['QuantityToken'] = c['QuantityToken'].map(token_to_unit)
    c['Year'] = int(file.split('_')[3])
    c['Type'] = file.split('_')[4]
    c['ProductType'] = c['ProductCode'].str.split('-', 1).str[1].str.strip()
    c['ProductCode'] = c['ProductCode'].str.split('-', 1).str[0].str.strip()
    c = c[c['ProductCode'].str.contains('PDR|WHT|GRO|V_F|OSD|CTL|OAP|FRS|CMT|OMT|LUM')]
    c.columns = [
        'Country',
        'Product Code',
        'Quantity Unit',
        'Value (millions)',
        'Quantity',
        'Weight',
        'Year',
        'Type',
        'Product Type',
    ]
    c = c.reindex(
        columns=[
            'Country',
            'Year',
            'Type',
            'Product Code',
            'Product Type',
            'Weight',
            'Quantity',
            'Quantity Unit',
            'Value (millions)',
        ]
    )
    trade.append(c)

trade = pd.concat(trade)
trade = trade[trade['Type'] == 'Export']
trade = trade[~trade['Quantity Unit'].str.contains('N.A.|N.Q.')]

# display(trade[~trade['Country'].isin(dataset['Country'].unique())]['Country'].unique())
# display(dataset[~dataset['Country'].isin(trade['Country'].unique())]['Country'].unique())

trade_alignment = {
    'Anguila': 'Anguilla',
    'Bahamas, The': 'Bahamas',
    'Czech Republic': 'Czechia',
    'Egypt, Arab Rep.': 'Egypt',
    'Ethiopia(excludes Eritrea)': 'Ethiopia',
    'Faeroe Islands': 'Faroe Islands',
    'Gambia, The': 'Gambia',
    'Hong Kong, China': 'Hong Kong',
    'Iran, Islamic Rep.': 'Iran',
    'Kyrgyz Republic': 'Kyrgyzstan',
    'St. Kitts and Nevis': 'Saint Kitts and Nevis',
    'Korea, Rep.': 'South Korea',
    'St. Lucia': 'Saint Lucia',
    'Macedonia, FYR': 'Macedonia',
    'Russian Federation': 'Russia',
    'Slovak Republic': 'Slovakia',
    'St. Vincent and the Grenadines': 'Saint Vincent & Grenadines',
    'Syrian Arab Republic': 'Syria',
    'Micronesia, Fed. Sts.': 'Micronesia',
    'East Timor': 'Timor-Leste',
    'Congo, Rep.': 'Congo',
    'Occ.Pal.Terr': 'Palestine',
    'Lao PDR': 'Laos',
    'Turks and Caicos Isl.': 'Turks and Caicos Islands',
    'Cote d'Ivoire': 'Côte d’Ivoire',
}

trade.dropna(inplace=True)
trade['Country'].replace(trade_alignment, inplace=True)
trade.to_csv("datasets/processed/trade.csv", index=False)

dataset = pd.merge(dataset, trade, how='outer', on=['Country', 'Year'])
````

### Protected Area

Datasets sources list: No.7

````python
%%capture --no-display

# !pip install floweaver (under 2.0.0)
# !pip install ipysankeywidget
# $jupyter nbextension enable --py --sys-prefix ipysankeywidget
# $jupyter nbextension enable --py --sys-prefix widgetsnbextension

dataset = pd.read_csv("datasets/processed/dataset.csv")


def categorize_to_class(data_list):
    return pd.cut(
        x=data_list,
        bins=[
            min(data_list),
            np.quantile(data_list, q=0.25),
            np.quantile(data_list, q=0.5),
            np.quantile(data_list, q=0.75),
            max(data_list),
        ],
        labels=['class1', 'class2', 'class3', 'class4'],
    )


top_deforestation = (
    dataset.groupby(['Country'], as_index=False)[
        ['Total Deforestation Area (hectares)']
    ]
    .sum()
    .sort_values(by='Total Deforestation Area (hectares)', ascending=False)
    .set_index('Country')
    .head(50)
    .index.tolist()
)

deforestation_countries_export = dataset.dropna(
    subset=['Total Deforestation Area (hectares)', 'Product Type']
)
deforestation_countries_export = deforestation_countries_export[
    ~deforestation_countries_export['Quantity Unit'].str.contains('Item|m³|Liter')
]
deforestation_countries_export = deforestation_countries_export[
    deforestation_countries_export['Country'].isin(top_deforestation)
]
deforestation_stats = deforestation_countries_export.describe()
deforestation_countries_export = deforestation_countries_export[
    [
        'Country',
        'Year',
        '% Deforestation',
        'Product Type',
        'Quantity Unit',
        'Value (millions)',
    ]
].copy()

deforestation_countries_export.to_csv(
    "datasets/processed/deforestation_countries_export.csv", index=False
)

deforestation_countries_export_2000 = deforestation_countries_export[
    deforestation_countries_export['Year'] == 2000
]
deforestation_countries_export_2000.loc[
    :, 'Deforestation Category'
] = categorize_to_class(deforestation_countries_export_2000['% Deforestation'].values)
deforestation_countries_export_2000 = deforestation_countries_export_2000.groupby(
    ['Country', 'Deforestation Category', 'Product Type'], as_index=False
)['Value (millions)'].sum()
deforestation_countries_export_2000.rename(
    columns={
        'Country': 'source',
        'Deforestation Category': 'type',
        'Value (millions)': 'value',
        'Product Type': 'target',
    },
    inplace=True,
)


deforestation_countries_export_2010 = deforestation_countries_export[
    deforestation_countries_export['Year'] == 2010
]
deforestation_countries_export_2010.loc[
    :, 'Deforestation Category'
] = categorize_to_class(deforestation_countries_export_2010['% Deforestation'].values)
deforestation_countries_export_2010 = deforestation_countries_export_2010.groupby(
    ['Country', 'Deforestation Category', 'Product Type'], as_index=False
)['Value (millions)'].sum()
deforestation_countries_export_2010.rename(
    columns={
        'Country': 'source',
        'Deforestation Category': 'type',
        'Value (millions)': 'value',
        'Product Type': 'target',
    },
    inplace=True,
)


deforestation_countries_export_2015 = deforestation_countries_export[
    deforestation_countries_export['Year'] == 2015
]
deforestation_countries_export_2015.loc[
    :, 'Deforestation Category'
] = categorize_to_class(deforestation_countries_export_2015['% Deforestation'].values)
deforestation_countries_export_2015 = deforestation_countries_export_2015.groupby(
    ['Country', 'Deforestation Category', 'Product Type'], as_index=False
)['Value (millions)'].sum()
deforestation_countries_export_2015.rename(
    columns={
        'Country': 'source',
        'Deforestation Category': 'type',
        'Value (millions)': 'value',
        'Product Type': 'target',
    },
    inplace=True,
)


from floweaver import Bundle, Partition, ProcessGroup, SankeyDefinition, Waypoint, weave

size = dict(width=1000, height=500)

nodes = {
    'Country': ProcessGroup(top_deforestation, title='Country'),
    'Product Type': ProcessGroup(
        [
            'Animal products nec',
            'Bovine cattle, sheep and goats, horses',
            'Bovine meat products',
            'Cereal grains nec',
            'Forestry',
            'Meat products nec',
            'Oil seeds',
            'Paddy rice',
            'Vegetables, fruit, nuts',
            'Wheat',
            'Wood products',
        ],
        title='Export Product Type',
    ),
}

country_with_others = Partition.Simple(
    'process',
    [
        'Brazil',
        'India',
        'Indonesia',
        'Tanzania',
        'Australia',
        'Paraguay',
        'South Africa',
        'China',
        'Mexico',
        'Argentina',
        ('other', top_deforestation[10:]),
    ],
)

product_type = Partition.Simple(
    'process',
    [
        'Animal products nec',
        'Bovine cattle, sheep and goats, horses',
        'Bovine meat products',
        'Cereal grains nec',
        'Forestry',
        'Meat products nec',
        'Oil seeds',
        'Paddy rice',
        'Vegetables, fruit, nuts',
        'Wheat',
        'Wood products',
    ],
)

deforestation_category = Partition.Simple(
    'type', ['class1', 'class2', 'class3', 'class4']
)

nodes['Country'].partition = country_with_others
nodes['Product Type'].partition = product_type
nodes['waypoint'] = Waypoint(deforestation_category, title='Deforestation Category')

ordering = [
    ['Country'],
    ['waypoint'],
    ['Product Type'],
]

bundles = [
    Bundle('Country', 'Product Type', waypoints=['waypoint']),
]

sdd = SankeyDefinition(nodes, bundles, ordering, flow_partition=deforestation_category)

display(
    weave(sdd, deforestation_countries_export_2000)
    .to_widget(**size)
    .auto_save_png(filename="deforestation_countries_export_2000")
)
display(
    weave(sdd, deforestation_countries_export_2010)
    .to_widget(**size)
    .auto_save_png(filename="deforestation_countries_export_2010")
)
display(
    weave(sdd, deforestation_countries_export_2015)
    .to_widget(**size)
    .auto_save_png(filename="deforestation_countries_export_2015")
)
````

````python
%%capture --no-display


def categorize_to_class(data_list):
    return pd.cut(
        x=data_list,
        bins=[
            min(data_list),
            np.quantile(data_list, q=0.25),
            np.quantile(data_list, q=0.5),
            np.quantile(data_list, q=0.75),
            max(data_list),
        ],
        labels=['class1', 'class2', 'class3', 'class4'],
    )


deforestation_countries_export = pd.read_csv(
    "../datasets/processed/deforestation_countries_export.csv"
)

deforestation_countries_export_2015 = deforestation_countries_export[
    deforestation_countries_export['Year'] == 2015
]
deforestation_countries_export_2015.loc[
    :, 'Deforestation Category'
] = categorize_to_class(deforestation_countries_export_2015['% Deforestation'].values)
deforestation_countries_export_2015 = deforestation_countries_export_2015.groupby(
    ['Country', 'Deforestation Category', 'Product Type'], as_index=False
)['Value (millions)'].sum()
deforestation_countries_export_2015.rename(
    columns={
        'Country': 'source',
        'Deforestation Category': 'type',
        'Value (millions)': 'value',
        'Product Type': 'target',
    },
    inplace=True,
)

selected_c = [
    'Brazil',
    'India',
    'Indonesia',
    'Tanzania',
    'Paraguay',
    'South Africa',
    'China',
    'Mexico',
    'Argentina',
]
countries_other_dict = dict()
for c in deforestation_countries_export_2015[
    ~deforestation_countries_export_2015['source'].isin(selected_c)
]['source'].unique():
    countries_other_dict[c] = 'Others'

deforestation_countries_export_2015['source'] = deforestation_countries_export_2015[
    'source'
].replace(countries_other_dict)
deforestation_countries_export_2015.to_csv(
    "latest_deforestation_countries_export.csv", index=False
)
````

````python
num_of_rows = 10
num_of_columns = 10
num_of_countries = num_of_rows * num_of_columns

dataset = pd.read_csv("../datasets/processed/dataset.csv")
dataset = dataset[
    ~dataset['Country'].str.contains('Melanesia|Greenland|Côte d’Ivoire|Western Sahara')
]
top_countries_by_land_forest = (
    dataset.fillna(0)
    .sort_values(
        by=['Total Land Area (hectares)', '% Forest Cover Area'], ascending=False
    )['Country']
    .unique()[:num_of_countries]
)

scaled_dataset = pd.read_csv("../datasets/processed/scaled_dataset.csv")
forest_area_to_output = scaled_dataset[
    [
        'Country',
        'Year',
        '% Forest Cover Area',
        'Crop Output Quantity',
        'Agricultural Land Quantity',
    ]
]
forest_area_to_output = forest_area_to_output[
    forest_area_to_output['Country'].isin(top_countries_by_land_forest)
]
forest_area_to_output = forest_area_to_output.drop_duplicates(
    subset=['Country', 'Year']
)
# forest_area_to_output = forest_area_to_output.fillna(0)

rows_list = list()
columns_list = list()

for i in range(num_of_rows):
    for j in range(num_of_columns):
        rows_list.append(i)
        columns_list.append(j)

countries_position = dict()
counter = 0
for c in top_countries_by_land_forest:
    countries_position[c] = str(rows_list[counter]) + ',' + str(columns_list[counter])
    counter += 1

forest_area_to_output['Pos'] = forest_area_to_output['Country'].map(countries_position)
forest_area_to_output[['Rows', 'Columns']] = forest_area_to_output['Pos'].str.split(
    ',', 1, expand=True
)
forest_area_to_output.drop(columns='Pos', inplace=True)
forest_area_to_output.to_csv("forest_area_to_output.csv", index=False)
````
